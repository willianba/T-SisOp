```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
thr2_100 = []
thr2_200 = []
thr2_300 = []
thr2_400 = []
thr2_500 = []
thr2_600 = []
thr2_700 = []
thr2_800 = []
thr2_900 = []
thr2_1k  = []
thr2_1k1 = []
thr2_1k2 = []
thr2_1k3 = []
thr2_1k4 = []
thr2_1k5 = []
thr2_1k6 = []
thr2_1k7 = []
thr2_1k8 = []
thr2_1k9 = []
thr2_2k  = []
```

```python
arrays = [ 
    thr2_100, thr2_200, thr2_300, thr2_400, thr2_500, thr2_600, thr2_700, thr2_800, thr2_900, thr2_1k,
    thr2_1k1, thr2_1k2, thr2_1k3, thr2_1k4, thr2_1k5, thr2_1k6, thr2_1k7, thr2_1k8, thr2_1k9, thr2_2k
]
```

```python
os.chdir("../../tempos_paralelo")
```

```python
times = []
files = os.listdir(os.getcwd())

for file in os.listdir(os.getcwd()):
    with open(file) as f:
        times.append([x.replace("\n", "") for x in f.readlines()])
```

```python
for size in range(10):
    for num in range(20):
        arrays[num].append(float(times[size][num]))
```

```python
min_values = [min(thr2_100), min(thr2_200), min(thr2_300), min(thr2_400), min(thr2_500), min(thr2_600), min(thr2_700), min(thr2_800), min(thr2_900), min(thr2_1k),
              min(thr2_1k1), min(thr2_1k2), min(thr2_1k3), min(thr2_1k4), min(thr2_1k5), min(thr2_1k6), min(thr2_1k7), min(thr2_1k8), min(thr2_1k9), min(thr2_2k)]
avr_values = [np.average(thr2_100), np.average(thr2_200), np.average(thr2_300), np.average(thr2_400), np.average(thr2_500), np.average(thr2_600), np.average(thr2_700), np.average(thr2_800), np.average(thr2_900), np.average(thr2_1k),
              np.average(thr2_1k1), np.average(thr2_1k2), np.average(thr2_1k3), np.average(thr2_1k4), np.average(thr2_1k5), np.average(thr2_1k6), np.average(thr2_1k7), np.average(thr2_1k8), np.average(thr2_1k9), np.average(thr2_2k)]
max_values = [max(thr2_100), max(thr2_200), max(thr2_300), max(thr2_400), max(thr2_500), max(thr2_600), max(thr2_700), max(thr2_800), max(thr2_900), max(thr2_1k),
              max(thr2_1k1), max(thr2_1k2), max(thr2_1k3), max(thr2_1k4), max(thr2_1k5), max(thr2_1k6), max(thr2_1k7), max(thr2_1k8), max(thr2_1k9), max(thr2_2k)]
```

```python
fig, axes = plt.subplots(figsize=(12,5))

axes.plot(min_values, label="min", color="green")
axes.plot(avr_values, label="avr", color="blue")
axes.plot(max_values, label="max", color="red")
axes.set_ylabel("Tempo de execução (s)")
axes.set_xlabel("Tamanho matrizes")
axes.set_xticks(range(0,20))
axes.set_xticklabels(range(100,2100,100))
axes.set_title("2 Threads")

# custom grid appearance
axes.grid(color='black', alpha=0.4, linestyle='dashed', linewidth=0.5)

axes.legend()
fig.savefig("paralelo_2threads_titulo.jpg", dpi=200)
```
